```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.fft import fft, ifft, fftshift
from scipy.io.wavfile import read
import sounddevice as sd

```

```python
#A.1
N = 10
n = np.arange(N)
x1 = np.exp(1j*2*np.pi*n*30/100) + np.exp(1j*2*np.pi*n*33/100)
x2 = np.cos(2*np.pi*n*30/100) + 0.5*np.cos(2*np.pi*n*40/100)
X1 = fft(x1)
X2 = fft(x2)
f_r = np.linspace(-0.5, 0.4, N)

plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.stem(n, np.real(x1))
plt.title('Real part of x1')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 2)
plt.stem(n, np.imag(x1))
plt.title('Imaginary part of x1')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 3)
plt.stem(n, x2)
plt.title('Signal x2')
plt.xlabel('n')
plt.ylabel('Amplitude')


plt.figure(figsize=(14, 8))
plt.subplot(2, 1, 1)
plt.stem(f_r, np.abs(fftshift(X1)))
plt.title('Magnitude of FFT of x1')
plt.xlabel('Frequency (f_r)')
plt.ylabel('Magnitude')

plt.subplot(2, 1, 2)
plt.stem(f_r, np.abs(fftshift(X2)))
plt.title('Magnitude of FFT of x2')
plt.xlabel('Frequency (f_r)')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()
```

```python
#A.2
N = 500
x1_zp = np.append(x1, np.zeros(490))
x2_zp = np.append(x2, np.zeros(490))
X1_zp = fft(x1_zp)
X2_zp = fft(x2_zp)
f_r = np.linspace(-0.5, 0.498, N)

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(x1_zp)
plt.title("Zero-padded x1 signal")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")

plt.subplot(2, 2, 2)
plt.plot(x2_zp)
plt.title("Zero-padded x2 signal")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")

plt.subplot(2, 2, 3)
plt.plot(f_r, np.abs(fftshift(X1_zp)) / N)
plt.title("Magnitude of X1_zp")
plt.xlabel("Normalized Frequency")
plt.ylabel("Magnitude")

plt.subplot(2, 2, 4)
plt.plot(f_r, np.abs(fftshift(X2_zp)) / N)
plt.title("Magnitude of X2_zp")
plt.xlabel("Normalized Frequency")
plt.ylabel("Magnitude")

plt.tight_layout()
plt.show()
```

```python
#A.3
N = 100
n = np.arange(N)
x1 = np.exp(1j*2*np.pi*n*30/100) + np.exp(1j*2*np.pi*n*33/100)
x2 = np.cos(2*np.pi*n*30/100) + 0.5*np.cos(2*np.pi*n*40/100)
X1 = fft(x1)
X2 = fft(x2)
f_r = np.linspace(-0.5, 0.49, N)

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(n, np.real(x1))
plt.title('Real Part of x1')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 2)
plt.plot(n, x2)
plt.title('Real Part of x2')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 3)
plt.plot(f_r, np.abs(fftshift(X1)))
plt.title('Magnitude of X1')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')

plt.subplot(2, 2, 4)
plt.plot(f_r, np.abs(fftshift(X2)))
plt.title('Magnitude of X2')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.tight_layout()
plt.show()
```

```python
#A.4
N = 500
x1_zp = np.append(np.append(np.zeros(200), x1), np.zeros(200))
x2_zp = np.append(np.append(np.zeros(200), x2), np.zeros(200))
X1_zp = fft(x1_zp)
X2_zp = fft(x2_zp)
f_r = np.linspace(-0.5, 0.498, N)

plt.figure(figsize=(14, 8))
plt.subplot(2, 2, 1)
plt.plot(x1_zp)
plt.title('Zero-Padded x1')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 2)
plt.plot(x2_zp)
plt.title('Zero-Padded x2')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 3)
plt.plot(f_r, np.abs(fftshift(X1_zp)))
plt.title('Magnitude Spectrum of X1_zp')
plt.xlabel('Normalized Frequency')
plt.ylabel('Magnitude')

plt.subplot(2, 2, 4)
plt.plot(f_r, np.abs(fftshift(X2_zp)))
plt.title('Magnitude Spectrum of X2_zp')
plt.xlabel('Normalized Frequency')
plt.ylabel('Magnitude')
plt.tight_layout()
plt.show()
```
