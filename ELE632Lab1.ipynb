```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#A-1-i
x = np.arange(-5,5)

def y(n):
    unit = np.array([])
    for sample in n:
        if sample == 3:
            unit = np.append(unit,1)
        else:
            unit = np.append(unit,0)
    return(unit)

plt.figure()
plt.stem(x,y(x))
plt.title('δ[n-3]')
plt.show()
```

```python
#A-1-ii
x= np.arange(-10,10,0.5)

def u(a, n):
    unit = np.array([])
    for sample in n:
        if sample<a:
            unit = np.append(unit,0)
        else:
            unit = np.append(unit,1)
    return(unit)

plt.figure()
plt.stem(x, u(-1,x))
plt.title('u[n+1]')
plt.show()
```

```python
#A-1-iii
x = np.arange(-10,25,0.5)

def u(a, n):
    unit = np.array([])
    for sample in n:
        if sample<a:
            unit = np.append(unit,0)
        else:
            unit = np.append(unit,np.cos(np.pi*0.2*sample))
    return(unit)

plt.figure()
plt.stem(x, u(0,x))
plt.title('x[n] = cos(πn/5)u[n]')
plt.show()
```

```python
#C-1
x = np.zeros(99)
for n in range(1, 100):
    x1 = np.cos(np.pi * n / 5)
    x2 = 1 if n - 30 == 0 else 0
    x3 = 1 if n - 35 == 0 else 0
    x[n-1] = x1 + x2 - x3

# Filter Signal Design
B4 = np.zeros_like(x)  # Padding input vector for 4 point filter
B8 = np.zeros_like(x)  # Padding input vector for 8 point filter
B12 = np.zeros_like(x)  # Padding input vector for 12 point filter
modifyx = x.copy()

N = 4  # 4 point filter
for i in range(len(modifyx) - N):
    window = x[i:i+N]  # cut the element specified by N
    B4[i] = np.max(window)  # filter out the max component

N = 8  # 8 point filter
for i in range(len(modifyx) - N):
    window = x[i:i+N]  # cut the element specified by N
    B8[i] = np.max(window)  # filter out the max component

N = 12  # 12 point filter
for i in range(len(modifyx) - N):
    window = x[i:i+N]  # cut the element specified by N
    B12[i] = np.max(window)  # filter out the max component


plt.figure(figsize=(12, 8))
plt.subplot(4, 1, 1)
plt.stem(x)
plt.title('Original Signal')

plt.subplot(4, 1, 2)
plt.stem(B4)
plt.title('4-point Filter')

plt.subplot(4, 1, 3)
plt.stem(B8)
plt.title('8-point Filter')

plt.subplot(4, 1, 4)
plt.stem(B12)
plt.title('12-point Filter')
plt.show()
```

```python
#D1
def energy_power(x):
    energy = np.sum(np.abs(x)**2)
    power = energy/len(x)
    return energy, power
x = np.array([1,2,3,4,5])

energy, power = energy_power(x)
print(f"Energy: {energy}")
print(f"Power: {power}")
```
